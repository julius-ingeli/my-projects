```python
bigL = []
newL = []
resD = {}


f = open("day1 calories.txt")
for x in f:    
    bigL.append(x[:-1])
    
    
itr = 1
for x in bigL:
    if x.isnumeric() == True:
        x = int(x)
        newL.append(x)

    else:
        elfSum = sum(newL)
        resD['Elf'+str(itr)]=elfSum
        newL=[]
        itr+=1
        
for key,val in resD.items():
    print(key,':',val)
```

```python
calL = []
for cal in resD.values():
    calL.append(cal)
```

```python
calL = sorted(calL)
print(calL)
```
