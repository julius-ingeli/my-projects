```python
initL = []
fhand = open('day2 data.txt','r')
for r in fhand:
    initL.append(r.replace("\n",""))
    
```

```python

yscore = 0
zscore = 0
for x in initL:
    y=0
    z = 0
    if x[0] == 'A': #rock
        y = 1
    if x[0] == 'B': #paper
        y = 2
    if x[0] == 'C': #scissors
        y = 3 
    if x[-1] == 'X': #rock
        z = 1
    if x[-1] == 'Y': #paper
        z = 2
    if x[-1] == 'Z': #scissors
        z = 3
        
    
    #print(x[0],y,z,x[-1])
    
    if (y==3 and z==3) or (y==2 and z==2) or (y==1 and z==1): #same thing
        yscore+=y+3
        zscore+=z+3
        print('draw!',end=' ')
    #scissor wins
    if z == 3 and y == 2: #player win
        yscore+=y
        zscore+=z+6
        print('player wins!',end=' ')
    if z == 2 and y == 3: #opponent win
        zscore+=z
        yscore+=y+6
        print('opponnent wins!',end=' ')
    #paper wins
    if z == 2 and y == 1:
        yscore+=y
        zscore+=z+6
        print('player wins!',end=' ')
    if z == 1 and y == 2:
        zscore+=z
        yscore+=y+6
        print('opponnent wins!',end=' ')
    #rock wins
    if z == 1 and y == 3:
        yscore+=y
        zscore+=z+6
        print('player wins!',end=' ')
    if z == 3 and y == 1:
        zscore+=z
        yscore+=y+6
        print('opponnent wins!',end=' ')
    
    
    print('oppnent score:',yscore,'player score:',zscore)
    
print("player score:",zscore)
print("opponent score:",yscore)
```
